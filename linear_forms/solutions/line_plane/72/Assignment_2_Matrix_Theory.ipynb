Find the equation of the line passing through the points
$\begin{pmatrix}
  3\\ 
  -2\\
  -5
\end{pmatrix}$  $\begin{pmatrix}
  3\\ 
  -2\\
  6
\end{pmatrix}$

__Solution__ : 


Let A = $\begin{pmatrix}
  3\\ 
  -2\\
  -5
\end{pmatrix}$ and B = $\begin{pmatrix}
  3\\ 
  -2\\
  6
\end{pmatrix}$

##### Python Code : 

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# assigning x,y,z co-ordinates and plotting in 3d
plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
# (x,y,z)Coordinates of given two points A and B
x_co = [3,3]
y_co= [-2,-2]
z_co = [-5,6]
result, = ax.plot3D(x_co, y_co, z_co,label="Line Passing through the points (3,-2,-5) and (3,-2,6)",color='r',marker='o')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
for x, y, z in zip(x_co, y_co, z_co):
    label = '(%d, %d, %d)' % (x, y, z)
    ax.text(x, y, z, label)
plt.legend(handles=[result])
```
