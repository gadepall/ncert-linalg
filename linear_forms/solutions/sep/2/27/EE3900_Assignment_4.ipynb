```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#assigning normals of Planes 
normal_1 = np.array([[2], [2], [-3]])
c_1 = 7
normal_2 = np.array([[2], [5], [3]])
c_2 = 9

#assigning point through passes
A = np.array([[2],[1],[3]])

# plane which passes through intersection of P1 and P2 and passing through the point A
lamba = np.zeros((3,1))
normal = np.zeros((3,1))
lamba = (c_1-normal_1.T@A)/(normal_2.T@A-c_2)
normal = normal_1 +lamba*(normal_2)

c = c_1+lamba*c_2
print('Value of lambda is',lamba)
print('Normal of The plane $P$ is \n ',normal)
# Defining X and Y for plotting
xx, yy = np.meshgrid(range(-20, 20), range(-20, 20))

# Equation of Z corresponding X and Y
#for planes P1 ans P2
zz_1 = ((c_1) - (normal_1[0] * xx) - (normal_1[1] * yy))/(normal_1[2])
zz_2 = ((c_2) - (normal_2[0] * xx) - (normal_2[1] * yy))/(normal_2[2])
#for plane P
zz = ((c) - (normal[0] * xx) - (normal[1] * yy))/(normal[2])

#setting up plot
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

#plotting planes P1,P2 and P
Plane1=ax.plot_surface(xx,yy,zz_1, color='b',alpha=0.25,label='(2 2 -3)x = 7')
Plane1._facecolors2d=Plane1._facecolors3d
Plane1._edgecolors2d=Plane1._edgecolors3d

Plane2=ax.plot_surface(xx,yy,zz_2, color='b',alpha=0.25,label='(2 5 3)x = 9')
Plane2._facecolors2d=Plane2._facecolors3d
Plane2._edgecolors2d=Plane2._edgecolors3d

Plane=ax.plot_surface(xx,yy,zz, color='y',alpha=0.25,label='Plane $P$')
Plane._facecolors2d=Plane._facecolors3d
Plane._edgecolors2d=Plane._edgecolors3d

plt.xlabel('$x$');plt.ylabel('$y$')
plt.grid()
plt.legend()

plt.show()
```
